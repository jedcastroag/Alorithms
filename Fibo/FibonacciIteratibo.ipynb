# Fibonacci Iterativo - Análiss de Tiempo de Ejecución

Librerias necesarias para gráficar y tomar el tiempo de ejecución:

```python
import time as t
import numpy as np
import matplotlib.pyplot as plt
```

Los siguientes parametro son: 'N', el N-esimo número al que se le calculará el fibonacci y el tiempo de ejecución; e 'inc', el incremento que vamos a hacer desde cero hasta N.

```python
N = 100000
inc = 500
```

### Implementación

```python
def fibo(n):
    if(n < 1):
        return 0
    f0 = 0
    f1 = 1
    for x in range(1,n):
        temp = f0 + f1
        f0 = f1
        f1 = temp
    return f1
```

### Calculo de los Tiempos de Ejecución

```python
def time_comp(n):   
    times = []
    for i in range(0,n+1,inc):
        start = t.time()
        fibo(i)
        finish = t.time()-start
        times.append([i,finish])
    return np.array(times)
```

### Grafica de los Tiempos de Ejecución 

```python
def plot(n):
    np_times = time_comp(n)
    plt.plot(np_times[:,0],np_times[:,1])
    plt.title("Tiempo_Ejecución(n)")
    plt.xlabel("N-esimo Termino de Fibonacci")
    plt.ylabel("Tiempo (s)")
```

Gráfica de Tiempos de Ejecución dependiendo del número de datos

```python
plot(N)
```
